```python
import numpy as np
import scipy.stats as stats

# Data
data = [25, 35, 40, 42, 45, 45, 45, 50, 52, 55, 60]

# Mean
mean = np.mean(data)
print(f"Mean: {mean:.2f}")

# Median
median = np.median(data)
print(f"Median: {median:.2f}")

# Mode
mode_result = stats.mode(data, keepdims=True)
print(f"Mode: {mode_result.mode[0]:.2f}")

# Weights (11 weights for 11 data values)
weights = [1, 1, 2, 3, 2, 4, 4, 4, 2, 1, 1]

# Weighted Mean
w_mean = np.average(data, weights=weights)
print(f"Weighted Mean: {w_mean:.2f}")
```
