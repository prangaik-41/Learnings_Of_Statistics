This Following coide related to the Detecting Outliers 
And Fences.

```python
import numpy as np

data = np.array([35,41,44,45,47,48,49,50,52,53,56,62,71,92])

q1,q3 = np.percentile(data,[25,75])
iqr=q3-q1
low,high  = q1-1.5*iqr,q3+1.5*iqr

outliers = data[(data<low)|(data>high)]
print("fences: ",low,high,"| Outliers: ",outliers)
```

This following code is  for Standard deviation and population and sample 

```python
import numpy as np 

marks = np.array([4,8,6,5,7])

print(np.var(marks))  #Numpy defaults to (ddof = 0) That is Population
print(np.std(marks))

print(np.var(marks,ddof=1))   #So we have to chnage  it to the (ddof=1) To Get the Sample
print(np.std(marks,ddof=1))
```
